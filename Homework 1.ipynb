```python
#підхід 2**h - 1
import math

def Pollard_mod_1(n, f, x_0):
    x = x_0[:]
    d = 0
    j = 0

    while d <= 1:
        j += 1
        print("\nj=", j)
        x_j = f(x[-1]) % n
        x.append(x_j)
        print(f"x_{j}=",x_j)
        h_cand = []
        for h in range(j+1):
            if (2**h <= j) and (2**(h+1) > j):
                h_cand.append(h)
        h = max(h_cand)
        print("h=", h)
        k = 2**h - 1
        print("k=", k)
        d = math.gcd(n, (x_j - x[k])%n) 
        if d > 1: 
            print(f"RETURN: (x_{k}, x{j})=({x[k]},{x_j}), gcd({n}, {(x_j - x[k])%n})={d}")
            return (x[k],x_j)
```

```python
#підхід j = 2k
import math

def Pollard_mod_2(n, f, x_0):
    x = x_0[:]
    d = 0
    j = 0

    while d <= 1:
        j += 1
        print("j=", j)
        x_j = f(x[-1]) % n
        x.append(x_j)
        print(f"x_{j}=",x_j)
        if j%2 == 0:
            k = int(j / 2)
            print("k=", k)
            d = math.gcd(n, (x_j - x[k])%n) 
            if d > 1: 
                print(f"RETURN: (x_{k}, x{j})=({x[k]},{x_j}),gcd({n}, {(x_j - x[k])%n})={d}")
                return (x[k],x_j)
```
